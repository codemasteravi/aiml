## Step 1: Import Required Libraries

```python
# Step 1: Import Required Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Step 2: Load and Preprocess the Dataset
The **Cats vs. Dogs dataset** contains labeled images of cats and dogs. We will:
- Download the dataset."https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip"
- Upload to /content
- Organize the images into training and validation directories.
- Preprocess the images using **resizing and normalization**.
- Apply **data augmentation** to improve model generalization.

```python
train_dir = "content/cats_and_dogs_filtered/train"
validation_dir = "content/cats_and_dogs_filtered/validation"
```

## Step 3: Apply Image Preprocessing and Data Augmentation
We apply:
- **Resizing**: Resize images to 150x150 pixels to maintain uniformity.
- **Data Augmentation**: Introduce variations such as rotation, flipping, zooming to make the model robust.
- **Normalization**: Scale pixel values between 0 and 1 for stable training.

```python
# Step 3: Apply Image Preprocessing and Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1.0/255)
print (train_dir)
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)
```

## Step 4: Build the CNN Model
We will define a **CNN model** with:
- **Convolutional layers** to extract features.
- **Max Pooling layers** to reduce dimensionality.
- **Dense layers** to perform classification.

```python
#TODO
# Step 4: Build the CNN Model

#TODO - compite the model with binary_crossentropy loss function
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

## Step 5: Train the CNN Model

```python
training = model.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator
)
```

# Step 6: Evaluate the Model

```python
# Step 6: Evaluate the Model
loss, accuracy = model.evaluate(validation_generator)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
